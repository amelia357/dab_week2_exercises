# Exercise 1: Lists and Dictionaries



```python
# Create a shopping list with at least 5 items

shopping_list = ["milk", "eggs", "bread", "bananas", "carrots"]
```

```python
print(shopping_list)
```

```python
# Print out the third item in the list

print(shopping_list[2])
```

```python
# Replace the first item with a different item

shopping_list = ["milk", "eggs", "bread", "bananas", "carrots"]

shopping_list.insert(0, "cereal")
shopping_list.pop(1)

print(shopping_list)
```

```python
# Add an item to the end of the list

shopping_list = ["cereal", "eggs", "bread", "bananas", "carrots"]
shopping_list.insert(5, "bacon")

print(shopping_list)
```

```python
# Create a dictionary of 3 grocery items with their prices

dict_groceries = {'cereal': 5.49, 'eggs': 3.49, 'bread': 2.00}

print(dict_groceries)
```

```python
# Print a message showing one item and its price

dict_groceries = {'cereal': 5.49, 'eggs': 3.49, 'bread': 2.00}

item, price = list(dict_groceries.items())[1]
print(item, price)
```

```python
# Add a new item to the dictionary with its price

dict_groceries = {'cereal': 5.49, 'eggs': 3.49, 'bread': 2.00}

dict_groceries["bacon"] = 6.49

print(dict_groceries)
```

```python
# Calculate and print the total cost of all items

dict_groceries = {'cereal': 5.49, 'eggs': 3.49, 'bread': 2.0, 'bacon': 6.49}
total = sum(dict_groceries.values())
print(total)

```
