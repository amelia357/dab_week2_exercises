```python
# Test Task 7
def celsius_to_fahrenheit(celsius):
    return celsius * 9/5 + 32

def fahrenheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9
```

```python
def class_celsius(celsius):
        if celsius >= 25:
            return "hot"
        elif celsius >= 15:
            return "moderate"
        elif celsius >= 1:
            return "cold"
        else:
            return "freezing"

def class_fahrenheit(fahrenheit):
        if fahrenheit >= 77:
            return "hot"
        elif fahrenheit >= 59:
            return "moderate"
        elif fahrenheit >= 32:
            return "cold"
        else:
            return "freezing"
```

```python
print(f"18°C is {celsius_to_fahrenheit(18)}°F and feels {class_fahrenheit(celsius_to_fahrenheit(18))}")
print(f"89°F is {fahrenheit_to_celsius(89)}°C and feels {class_celsius(fahrenheit_to_celsius(89))}")
print(f"26°C is {celsius_to_fahrenheit(26)}°F and feels {class_fahrenheit(celsius_to_fahrenheit(26))}")
print(f"32°F is {fahrenheit_to_celsius(32)}°C and feels {class_celsius(fahrenheit_to_celsius(32))}")
print(f"-12°C is {celsius_to_fahrenheit(-12)}°F and feels {class_fahrenheit(celsius_to_fahrenheit(-12))}")
print(f"10°F is {fahrenheit_to_celsius(10)}°C and feels {class_celsius(fahrenheit_to_celsius(10))}")
```

```python
# Task 9: Advanced Challenge - Student Grade Function
# Create a function named 'calculate_final_grade' that:
# 1. Takes a student's name and any number of assignment scores
# 2. Calculates the average score
# 3. Determines the letter grade based on the average (A: 90+, B: 80-89, C: 70-79, D: 60-69, F: <60)
# 4. Returns a formatted string with the student's name, average score, and letter grade
# Your code here


# Test Task 9
print("Task 9 Test:")
# Test with different student names and scores
# Print the results
print()
```

```python
# Test Task 9
dict_students = {'sue': [60, 75, 82, 40], 'joe': [95, 80, 83, 79], 'mary': [50, 45, 70, 65]} 

for keys, values in dict_students.items(): print(keys, values)
```

```python
def calculate_average(scores):
    return sum(scores) / len(scores)

def determine_letter_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 80:
        return "B"
    elif avg >= 70:
        return "C"
    elif avg >= 60:
        return "D"
    else:
        return "F"
```

```python
for name, scores in dict_students.items():
    avg = calculate_average(scores)
    grade = determine_letter_grade(avg)
    print(f"{name.title()} has an average of {avg:.2f} and earns a grade of {grade}")
```
